```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.integrate import quad

def payoff_function(v1, b1, b2, b3):
    """
    Calculate the payoff function for the first player in a second price auction game.

    Parameters:
    v1 (float): Value of the first player
    b1 (float): Bid of the first player
    b2 (float): Bid of the second player
    b3 (float): Bid of the third player

    Returns:
    float: Payoff for the first player
    """
    if b1 > max(b2, b3):
        return v1 - max(b2,b3)
    else:
        return 0

def expectation_function(v1, b1):
    """
    Calculate the mathematical expectation function for a given value of v1 and b1.

    Parameters:
    v1 (float): Value of the first player
    b1 (float): Bid of the first player

    Returns:
    float: Mathematical expectation
    """
    expectation = 0
    for b2 in np.linspace(0, 1, 100):
        for b3 in np.linspace(0, 1, 100):
            expectation += payoff_function(v1, b1, b2, b3) * 1/100 * 1/100
    return expectation

# Generate arbitrary values of v1
v1_values =np.arange(0.1,1,0.1)
b1_values=np.linspace(0, 1, 100)
# Calculate expectation values for each v1 and b1
expectation_values = np.zeros((9, 100))
for i, v1 in enumerate(v1_values):
    for j, b1 in enumerate(np.linspace(0, 1, 100)):
        expectation_values[i, j] = expectation_function(v1, b1)


# Find maximum value of mathematical expectation
max_expectation=[]
exp_val=[]
for i in range(0,len(expectation_values),1):
     max_expectation = np.max(expectation_values[i,:])
     exp_val.append(max_expectation)
     print(f"Maximum value of mathematical expectation: {max_expectation}")
idx_max=[]
for i in range(0,len(expectation_values),1):
  for j in range(0,np.max(np.shape(expectation_values)),1):
     if exp_val[i] == np.max(expectation_values[i,j]):
         idx_max.append(j)

# Find values of b1 at the maximum expectation
max_b1_values = b1_values[idx_max]
print(f"Values of b1 at maximum expectation: {max_b1_values}")

# Plot the expectation values
plt.figure(figsize=(12, 8))
for i in range(9):
    plt.plot(b1_values, expectation_values[i], label=f"v1={v1_values[i]}")
plt.xlabel('b1')
plt.ylabel('Expectation')
plt.plot(max_b1_values,exp_val, marker='o')
plt.legend()
plt.show()
# Plot b1 values in terms of v1 values
plt.figure(figsize=(8, 6))
plt.plot(max_b1_values,v1_values, marker='o')
plt.xlabel('v1')
plt.ylabel('b1 at max expectation')
plt.show()

# Fit a line function on the b1 values
coefficients = np.polyfit(v1_values, max_b1_values, 1)
line_function = np.poly1d(coefficients)
print(f"Line function: {line_function}")
```
